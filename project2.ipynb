## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/Web_scraping"  # 원하는 Wikipedia 페이지로 변경 가능

# 2️⃣ 웹페이지 요청 및 HTML 가져오기
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 3️⃣ HTML 태그 제거하고 텍스트만 추출
page_text = soup.get_text(separator="\n")  # 개행문자로 줄 구분

# 4️⃣ 텍스트 출력 (처음 500자만 미리보기)
print(page_text[:500])  

# 5️⃣ 파일로 저장 (선택 사항)
with open("wikipedia_text.txt", "w", encoding="utf-8") as file:
    file.write(page_text)
```

```python
#download an image with beautifulsoup and save it in this repository

import os
image_tag = soup.find("img")  # 첫 번째 <img> 태그 가져오기
if image_tag:
    image_url = "https://en.wikipedia.org/wiki/United_States#/media/File:Flag_of_the_United_States_(DoS_ECA_Color_Standard).svg" + image_tag["src"]  # 이미지 URL 완성
    image_name = os.path.basename(image_url)  # 파일명 추출
    
    # 4️⃣ 이미지 다운로드
    image_response = requests.get(image_url)
    
    # 5️⃣ 파일로 저장
    with open(image_name, "wb") as file:
        file.write(image_response.content)
    
    print(f"✅ Image downloaded and saved as {image_name}")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links

links = soup.find_all("a", href=True)  # 모든 <a> 태그에서 href 속성이 있는 것만 가져옴

# 4️⃣ 링크 URL 정리 및 출력 (10개만)
for i, link in enumerate(links[:10]):  # 상위 10개 링크만 출력
    link_url = link["href"]
    
    # Wikipedia 내부 링크인지 외부 링크인지 확인
    if link_url.startswith("/wiki/"):
        link_url = "https://en.wikipedia.org" + link_url  # 상대 경로를 절대 경로로 변환
    
    print(f"{i+1}. {link_url[:100]}") 
```

```python
def clean(text):

    #convert the text to lowercase
    text = text.lower()
    #remove punctuation
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    #remove newline and space characters
    text = ' '.join(text.split())
    return text

clean(soup.text)
```

```python
text_list = []
```

```python
for i, row in scripts.iterrows(): 
    link = row['link']
    
    if 'http' in str(link):  # 링크가 문자열인지 확인
        try: 
            source = requests.get(link, timeout=10)
            source.raise_for_status()  # HTTP 오류 체크
            
            try:
                soup = BeautifulSoup(source.text, "html.parser")
                text = soup.get_text(separator=" ").strip()[:1000]  # HTML 태그 제거 후 1000자만 저장
            except Exception as e:
                print(f"⚠️ Parsing error on {link}: {e}")
                text = ""  # 파싱 실패 시 빈 문자열 추가
            
        except requests.exceptions.RequestException as e:
            print(f"❌ Failed to fetch {link}: {e}")
            text = ""  # 요청 실패 시 빈 문자열 추가
    else:
        text = ""  # 유효하지 않은 링크인 경우 빈 문자열 추가

    text_list.append(text)  # 결과 리스트에 추가

# 📌 길이 맞추기
if len(text_list) < len(scripts):
    text_list.extend([""] * (len(scripts) - len(text_list)))  # 부족한 만큼 빈 문자열 추가
```

```python
#download the movie posters for 10 of these movies and save them to this repository

import requests
import pandas as pd
import os
from PIL import Image
from io import BytesIO

# 1️⃣ Load the dataset
movies_df = pd.read_csv("now_playing_movies.csv")

# 2️⃣ Create a folder to save images
save_folder = "movie_posters"
os.makedirs(save_folder, exist_ok=True)

# 3️⃣ Download the first 10 movie posters
num_movies = min(10, len(movies_df))  # If fewer than 10 movies are available, adjust
for i in range(num_movies):
    movie_title = movies_df.loc[i, "Title"]
    poster_url = movies_df.loc[i, "Poster URL"]

    if pd.isna(poster_url) or poster_url == "None":
        print(f"⚠️ No poster available for {movie_title}")
        continue  # Skip movies without posters

    try:
        # 🔹 Request the image
        response = requests.get(poster_url)
        response.raise_for_status()  # Check if the request was successful

        # 🔹 Open the image and save it
        image = Image.open(BytesIO(response.content))
        image_path = os.path.join(save_folder, f"{movie_title}.jpg")
        image.save(image_path, "JPEG")

        print(f"✅ Saved poster for {movie_title}")

    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to download {movie_title}: {e}")

print("🎬 All posters downloaded!")
```
